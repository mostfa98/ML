```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
img_as_array = mpimg.imread(r"C:\Users\mhija\Desktop\machine learning_course\DATA\palm_trees.jpg")
```

```python
img_as_array.shape #(H,W,C) C --> red blue green
```

```python
plt.imshow(img_as_array)
```

```python
(h,w,c) = img_as_array.shape
h
```

```python
# convert it into 2d because model only accept 2d input
Image_as_array2d = img_as_array.reshape(h*w,c)
```

```python
len(img_as_array.shape)
```

```python
len(Image_as_array2d.shape)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=6)
labels = model.fit_predict(Image_as_array2d)
```

```python
#this line is where we extract the representative colors (the quantized palette) from the k-means model
# If we kept all pixels in the cluster, we wouldnt be compressing anything , wed still have too many colors.
# that's why we use only the centers (mean of clusters)
#Quantizations goal is to reduce many similar colors into 1 representative color.

rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
qunatized_image = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
plt.imshow(qunatized_image)
```
