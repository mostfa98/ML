```python
def display (model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1',y='X2',hue=labels,palette='Set1')
    
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=3)
kpred = k_means.fit_predict(df)
display(k_means,df)
```

```python
# That's why you can't use k-means for such case it can't cluster two types 
# of data that are really close to each other 
k_means = KMeans(n_clusters=2)
kpred = k_means.fit_predict(moon)
display(k_means,moon)
```
