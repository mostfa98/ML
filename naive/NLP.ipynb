```python
with open('one.txt') as mytext:
    # read is the mood , lower and split reutern words only
    words_one = mytext.read().lower().split()
    # set return only the unique words no dublicate
    uni_words_one = set(words_one)
```

```python
uni_words_one
```

```python
with open('Two.txt') as mytext2:
    words_two = mytext2.read().lower().split()
    uni_words_two = set(words_two)
```

```python
all_uni_words = set()
all_uni_words.update(words_one)
all_uni_words.update(words_two)
# update like append :)
```

```python
full_vocab = dict()
i = 0
for word in all_uni_words:
    full_vocab[word] = i
    i = i + 1

full_vocab
```

```python
one_freq = [0]*len(full_vocab)
two_freq = [0]*len(full_vocab)
all_words = ['']*len(full_vocab)
```

```python
with open('one.txt') as f:
    one_text = f.read().lower().split()
```

```python
'''''''''
'full_vocab = {'this': 0, 'are': 1, 'dogs': 2}
one_text = ['dogs', 'are', 'dogs']
one_freq = [0, 0, 0]  # initial counts
Processing step-by-step:

"dogs" → index 2 → one_freq[2] += 1 → [0, 0, 1]

"are" → index 1 → one_freq[1] += 1 → [0, 1, 1]

"dogs" → index 2 → one_freq[2] += 1 → [0, 1, 2]
'''''

for word in one_text:
    word_ind = full_vocab[word]
    one_freq[word_ind] += 1
```

```python
one_freq
```

```python
with open('two.txt') as mewo:
    two_text = mewo.read().lower().split()

for word in two_text:
    word_ind = full_vocab[word]
    two_freq[word_ind] += 1
```

```python
two_freq
```

```python
for word in full_vocab:
    word_ind = full_vocab[word]
    all_words[word_ind] = word
```
