```python
text = ['This is a line',
        'This is another line',
        'Completly different line']
```

```python
from sklearn.feature_extraction.text import CountVectorizer ,TfidfTransformer

cv = CountVectorizer(stop_words='english')
cv.fit_transform(text)
```

```python
sparse_matrix =cv.fit_transform(text)
sparse_matrix.todense()
```

```python
cv.vocabulary_
```

```python
tfidf = TfidfTransformer()
```

```python
sparse_matrix
```

```python
results = tfidf.fit_transform(sparse_matrix)
```

```python
results.todense()
```

```python
# Does everything in one go:
# Tokenizes raw text (splits into words).
# Builds vocabulary (assigns indices to words).
# Counts words (term frequency matrix).
# Applies TF-IDF weighting.
# Input: Raw text (list of strings/documents).
# Output: TF-IDF matrix (sparse matrix form).
# If you start with raw text → use TfidfVectorizer (simpler, faster).
# If you already have counts or want modular control → use TfidfTransformer

from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer()
tv_results = tv.fit_transform(text)
tv_results.todense()
```
