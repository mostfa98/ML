```python
messy = pd.Series(['mostfa   ','bo;bo','    claira   '])
```

```python
messy.str.replace(';','').str.strip().str.capitalize()
```

```python
def cleanup (name):
    name = name.replace(';','')
    name = name.strip()
    name = name.capitalize()
    return name
```

```python
np.vectorize(cleanup)(messy)
```
