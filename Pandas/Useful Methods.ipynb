```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price <30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
df
```

```python
# regular FUN
def simple(num):
    return num*2
#simple Funcation
lambda num :num*2
```

```python
df['double'] = df['total_bill'].apply(lambda num:num*2)
```

```python
def quality (total , tip):
    if  tip/total > 0.25 :
        return 'Generous'
    else:
        return'other'
```

```python
""""" NOTE
One bracket (df['total_bill']) gives you a Series (1D column).

Two brackets (df[['total_bill', 'tip']]) gives you a DataFrame (2D table).

----------------------------------------------------------------------------------

.apply(..., axis=1)
.apply() lets you run a function across rows or columns.

axis=1 means apply the function row by row (across columns).

----------------------------------------------------------------------------------
 


"""
```

```python
df[['total_bill','tip']].apply(lambda df :quality(df['total_bill'],df['tip']),axis = 1)
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
