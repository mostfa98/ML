```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
def create_ages(mu = 50 ,sigma = 13 ,num_samples =100 ,seed=42):
    np.random.seed(seed)

    sample_age = np.random.normal(loc = mu,scale=sigma , size=num_samples)
    sample_age = np.round(sample_age,decimals=0)
    return sample_age
```

```python
sample = create_ages()
sample
```

```python
sns.displot(sample,bins=20)
```

```python
df = pd.read_csv('Ames_Housing_Data.csv')
pd.set_option('display.max_columns',None)  # Show all columns
df.sample(5)
```

```python
df[(df['Overall Qual']>8) & (df['SalePrice']<200000)]
```

```python
sns.scatterplot(x='Gr Liv Area' , y='SalePrice',data = df)
```

```python
drop_ind = df[(df['Gr Liv Area']>4000) & (df['SalePrice']<400000) ].index
drop_ind
```

```python
df = df.drop(drop_ind,axis=0)
```

```python
# remove the outliers which is in this case
#  houses with live area >4000 with really low price that less then 400000
# but you should make sure that nothing is ind
sns.scatterplot(x='Gr Liv Area' , y='SalePrice',data = df)
```
