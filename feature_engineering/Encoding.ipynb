```python
my_strings = df.select_dtypes(include='object')
my_numeric = df.select_dtypes(include='number')
```

```python
df_obj_dummies = pd.get_dummies(my_strings ,drop_first=True ).astype(int)
```

```python
final_df = pd.concat([my_numeric,df_obj_dummies ],axis=1)
pd.set_option('display.max_columns',None)
```
